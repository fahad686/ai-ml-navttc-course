```python
import pandas as pd
import numpy as np

#load dt set
df=pd.read_csv('customer_churn_dataset-testing-master.csv')

print(df.head()) #show first 5 rows
```

```python
# Quick overview
df.info()
```

```python
# Statistical summary (numeric only)
df.describe()
```

```python
# Check if dataset has missing values
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include=['int64', 'float64'])

num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
print(df.dtypes)
```

```python
# Fill numeric columns with median
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
num_cols.remove('Churn')  # Don't fill target column

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with 'Missing'
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
for col in cat_cols:
    df[col] = df[col].fillna('Missing')
```

```python
pip install  scikit-learn
```

```python
print(cat_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded = encoder.fit_transform(df[cat_cols])

encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(cat_cols))

# Merge encoded with numeric + target
final_df = pd.concat([df[num_cols], encoded_df, df['Churn']], axis=1)

final_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(final_df[num_cols])

scaled_df = pd.DataFrame(scaled, columns=num_cols)

# Replace numeric columns with scaled ones
for col in num_cols:
    final_df[col] = scaled_df[col]

final_df.head()
```
