```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('/Users/macbook/Desktop/AI/ai-ml-navttc-course/Assignments/assignment_6/Titanic-Dataset.csv')
```

```python
titanic.head()
```

```python
sns.countplot(x='class', data=titanic)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Passenger Count by Class')
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
sns.histplot(titanic['age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
sns.boxplot(x='class', y='fare', data=titanic)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
sns.countplot(x='embark_town', hue='survived', data=titanic)
plt.xlabel('Embarkation Town')
plt.ylabel('Passenger Count')
plt.title('Survival Count by Embark Town')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='class', data=titanic)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare Colored by Class')
plt.show()
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset('titanic')

# Create two subplots (side by side)
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Male passengers age distribution
sns.histplot(titanic[titanic['sex'] == 'male']['age'], bins=20, kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Age Distribution - Male')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')

# Female passengers age distribution
sns.histplot(titanic[titanic['sex'] == 'female']['age'], bins=20, kde=True, ax=axes[1], color='lightcoral')
axes[1].set_title('Age Distribution - Female')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
